### Dealing with Sex

```python
df_raw['sex'][df_raw.sex == 99] = np.median(df_raw['sex'].drop(85, axis = 0))
```

### Dealing with Age

```python
Medage = np.median(df_raw.age.drop(df_raw.index[df_raw.age >= 90], axis = 0))
df_raw.age[df_raw.age > 90] = Medage
```

#### Delete record with age < 10

```python
df_raw.drop(df_raw.index[df_raw.age <10], axis = 0, inplace=True)
```

### Dealing with Income

```python
df_raw['income'][df_raw.income > 6] = np.median(df_raw['income'].drop(df_raw.index[df_raw['income'] > 6], axis = 0))
```
